```python
import os
import torchaudio

waveform, sample_rate = torchaudio.load('../wake_words/data/Jun 8, 9.06 PMâ€‹.wav', normalize=False)
seg_duration = 6
start = 0
end = seg_duration * sample_rate
done = False
while not done:
    if end >= waveform.size(1):
        end = waveform.size(1)
        done = True

    seg = waveform[:, start:end]
    path = os.path.join('/home/hoang/Workspace/FRIDAY/models/wake_words/data/noise1', f'{start}_{end}.wav')
    torchaudio.save(path, seg, sample_rate)
    start = end
    end += seg_duration * sample_rate
```

```python

import torchaudio
import os

root = '/home/hoang/Workspace/FRIDAY/models/wake_words/data/noise1'
new_data = '/home/hoang/Workspace/FRIDAY/models/wake_words/data/noise2'
os.makedirs(new_data, exist_ok=True)

for filename in os.listdir(root):
    if filename.endswith('.wav'):
        path = os.path.join(root, filename)
        waveform, sample_rate = torchaudio.load(path, normalize=False)
        waveform  = waveform * 0.6
        path = os.path.join(new_data, filename)
        torchaudio.save(path, waveform, sample_rate)
```
